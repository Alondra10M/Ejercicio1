##GRAFICOS AVANZADOS

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Se selecciona solo las columnas numricas, de otra forma, genera error
numeric_columns = dataset.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap de correlacin')
plt.show()
```

otra opcion


```python
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap=sns.color_palette("rocket", as_cmap=True))
plt.title('Heatmap de correlacin')
plt.show()
```

```python
import matplotlib.colors as mcolors

custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom", ["blue", "white", "red"])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap=custom_cmap)
plt.title('Heatmap de correlacin')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm_r')
plt.title('Heatmap de correlacin')
plt.show()
```
